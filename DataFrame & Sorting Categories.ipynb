```python
#DataFrame 8
sort_order = ['Jan','Feb','Mar','Apr','May',
              'Jun','Jul','Aug','Sept','Oct','Nov','Dec']
data = {'month':{0:'Jan',1:'Feb',2:'Mar',3:'Apr',4:'May',5:'Jun',
            6:'Jul',7:'Aug',8:'Sept',9:'Oct',10:'Nov',11:'Dec'},
    'Sales' :{0:200,1:2130,2:345,3:340,4:486,5:697,
            6:465,7:648,8:378,9:658,10:789,11:358}
    }
df = pd.DataFrame(data)
df
```

```python
df.sort_values('month',ascending=True)
```

```python
df.sort_values('month',ascending=False)
```

```python
df.index = pd.CategoricalIndex(df['month'],categories=sort_order,ordered=True)
df
```

```python
df = df.sort_index().reset_index(drop=True)
df
```

```python
import matplotlib.pyplot as plt
plt.plot(df['month'],df["Sales"])
plt.show();
```

```python
data ={'Status':{0:'Good',1:"Poor",2:"Excellence",3:"Average",4:"Pass"},
       'Score':{0:60,1:30,2:70,3:50,4:40},
       'Students':{0:15,1:5,2:1,3:16,4:1}
      }
df =pd.DataFrame(data)
df
```

```python
sort_order = ['Excellence','Good','Average','Pass','Poor']
```

```python
df.index = pd.CategoricalIndex(df['Status'],categories=sort_order,ordered=True)
df
```

```python
df = df.sort_index().reset_index(drop=True)
df
```

```python
plt.bar(df['Status'],df['Students'],color='g');
```

```python
df_pie=df.set_index('Status')
plot = df_pie.plot.pie(y='Students', figsize=(5, 20))
plot;
```
