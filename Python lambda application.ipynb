```python
import pandas as pd
df=pd.DataFrame(
               {
                   "name":["Chandarana","Chandarana","Chandarana","Chandarana",
                                   "Chandarana","Chandarana","Chandarana","Chandarana","Chandarana"],
                   "attribute":["Sparkling Water_January","Still Water_January","Honey Badger_January","January_Total",
                       "Sparkling Water_February","Still Water_February","Honey Badger_February","February_Total","Growth_Februray"],
                    "value":[20,34,67,121, 45,67,90,202,0.66]
               }
               ) 
df
```

```python
#Step 1
def clean_extract(value):
    parts=value.split("_")
    if len(parts)==2:
        product,month=parts
    else:
        product,month=parts[0],None
        df.replace(" Total","").replace(" Growth","")
    return product,month
```

```python
#Step 2
df[['product','month']]=df['attribute'].apply(lambda x :pd.Series(clean_extract(x)))
df
```

```python
#Step 3
#Add Total column
df["Total"]= df.apply(lambda row : row['value'] if "Total" in row["attribute"] else None, axis=1)
#Add Growth column
df["Growth"] = df.apply(lambda row: row['value'] if "Growth" in row["attribute"] else None, axis= 1)
df
```

```python
#Step 4
df['product'] = df.apply(lambda row: "" if 'Total' in row['attribute'] or "Growth" in row['attribute'] else row['product'], axis=1)
df['month'] = df.apply(lambda row: "" if 'Total' in row['attribute'] or "Growth" in row['attribute'] else row['month'], axis=1)
df
```

```python
#Step 6
df['month'] = df['month'].ffill()
df
```

```python
#Step 7
df['month'] = df['month'].fillna(method='ffill')
df
```

```python
#Step 8
df_cleaned=df[['name','product','month','value','Total','Growth']]
df_cleaned
```

```python
# Manually forward-fill the 'month' column
prev_value = None
for i in range(len(df)):
    if pd.isna(df.loc[i, 'month']):
        df.loc[i, 'month'] = prev_value
    else:
        prev_value = df.loc[i, 'month']

df
```
