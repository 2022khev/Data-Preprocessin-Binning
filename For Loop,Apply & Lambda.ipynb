```python
#Iteritems cannot access specific columns
```

```python
#Using range is very effective as it access using iloc & loc indexing 
for i in range(len(df)):
    print(df.loc[i, "Species"], df.loc[i, "Sepal_Length"])
```

```python
#Using the index to access
for i in range(len(df)):
    print(df.iloc[i, 0], df.iloc[i, 1])
```

```python
#Using the itertuples and the function getattr()
for i in df.itertuples(index=True,name='Pandas'):
    print(getattr(i,"Species"), getattr(i,"Sepal_Length"))
```

```python
#iterrows()
#Adding new column with proportion of Sepal_Width
for i,j in df.iterrows():
    df.loc[i,"new"] = j["Sepal_Length"]/j["Sepal_Width"]*100
df
```

```python
#Range()
#Finding the percentage of Sepal_Length
for i in range(len(df)):
    df.loc[i,"Proportion"] = round((df.loc[i, "Sepal_Length"] /len(df)*100),3)
df
```

```python
#itertuples()
for i in df.itertuples(index=True,name='Pandas'):
    print(round(getattr(i,"Sepal_Width") /len(df)*100,2))
```

```python
# iterate through each row and select
# 'Species' and 'Sepal_length' column respectively.
for row in df.itertuples(index=True, name='Pandas'):
    print(getattr(row, "Sepal_Width"), getattr(row, "Sepal_Length"))
```
